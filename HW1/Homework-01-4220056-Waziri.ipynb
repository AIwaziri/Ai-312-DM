```python
# Function to calculate the median from grouped data
def grouped_median(classes, frequencies):
    import numpy as np

    # Calculate cumulative frequencies
    cumulative_frequencies = np.cumsum(frequencies)

    # Total number of observations
    N = sum(frequencies)

    # Find median class (first class where CF >= N/2)
    median_position = N / 2
    for i, cf in enumerate(cumulative_frequencies):
        if cf >= median_position:
            median_class_index = i
            break

    # Extract values needed for the formula
    L = classes[median_class_index][0]  # Lower boundary of median class
    CF = cumulative_frequencies[median_class_index - 1] if median_class_index > 0 else 0
    f_m = frequencies[median_class_index]  # Frequency of median class
    h = classes[median_class_index][1] - classes[median_class_index][0]  # Class width

    # Apply the median formula
    median = L + ((median_position - CF) / f_m) * h
    return median
```

```python
class_intervals = [(0, 20), (20, 40), (40, 60), (60, 80), (80, 100)]
frequencies = [4, 12, 9, 20, 15]
```

```python
# Calculate the median
median_value = grouped_median(class_intervals, frequencies)
print("The median of the grouped data is:", median_value)
```

```python
plt.figure(figsize=(7, 5))
intervals = [str(interval) for interval in class_intervals]
plt.bar(intervals, frequencies, color="salmon", alpha=0.7)
plt.axhline(y=30, color="black", linestyle="--", label="Median Position (N/2)")
plt.title("Histogram of Grade Distribution")
plt.xlabel("Grade Intervals")
plt.ylabel("Frequency (# of Students)")
plt.legend()
plt.show()
```
